# La interfaz de usuario

Es el medio por el que el usuario se comunica con la máquina utilizando para ello los [dispositivos de E/S](https://colab.research.google.com/drive/1WG4hFI0tdwGCDH-wkM1SmDYR3tlby4Pz).

En un coche, la interfaz de usuario son los mandos del coche, a saber, volante, pedales, cambio y todos los indicadores y botones y palancas de la consola. A través de ella observamos y controlamos el estado de la máquina y también por medio de ella podemos interactuar con la máquina para que se comporte como necesitamos.

De igual modo nuestros programas, especialmente cuanto más complejos son, necesitarán de una interfaz de usuario para que este interáctue con el software y obtener así el resultado esperado.

## Tipos de interfaces de usuario

*  **CLI**: Interfaces de linea de comandos. El control del programa se hace en base a comandos
*  **Alfanuméricas**: Interfaces realizadas en formato texto
*  **GUI**: Interfaces gráficas, más rápidas e intuitivas. También son más complejas y se realizan con graficos y texto.
* **Naturales**: Táctiles, sensibles a los gestos o al lenguaje. Son más modernas

En este curso veremos las tres primeras. La interfaz de comandos la usamos cada vez que ejecutamos un script desde la consola con el comando `python` y el nombre de nuestro script como argumento

Las alfanuméricas las desarrollaremos en los primeros módulos para ver la problemática específica del desarrollo de interfaces de usuario
