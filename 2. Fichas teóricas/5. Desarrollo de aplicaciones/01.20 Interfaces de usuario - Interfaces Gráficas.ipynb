# Interfaces gráficas

Las GUI o interfaces gráficas son un programa informático que realiza la función de interfaz de usuario. Está formado por imágenes y objetos gráficos, que representan la información y acciones que se encuentran en la interfaz. Su objetivo es el de crear un entorno visual fácil de usar para que fluya la comunicación son el sistema operativo.

Lo más común es que las acciones sean de manipulación directa, para simplificar la interacción y mejorar la comunicación entre usuario y sistema operativo. Algunas GUI de ejemplo son los entornos de escritorio de Windows, el X-Window de GNU/Linux o el Aqua, de Mac OS X.

En el proceso de creación de interfaces gráficas de usuarios el desarrollador de aplicaciones multiplataforma es el encargado de crearlas.

## Programación orientada a eventos

Las interfaces de usuario pretenden permitir que el usuario pueda realizar diferentes acciones sobre el sistema. En las interfaces de texto esto se logra con comandos de consola, en las interfaces gráficas se pretende que respondan a la acción del usuario sobre ciertos objetos gráficos (botones, cajas de texto,...). Estas acciones se denominan eventos y la programación orientada a eventos es una forma de programación en la que tanto la estructura como la ejecución de los programas van determinados por los sucesos que ocurran en el sistema, definidos por el usuario o que ellos mismos provoquen.

## Interfaces gráficas en python

Uno de los estándares de creación de interfaces gráficas de usuario para python es [tkinter](https://docs.python.org/3/library/tk.html). Viene instalado por defecto en las distribuciones estandar de python para windows, mac y la mayoría de distribuciones para Linux. A partir de la versión 8, tkinter ha mejorado lo bastante como para realizar apps de escritorio en estos tres SO.

Para consultar si tenemos tkinter en nuestra versión de python se puede ejecutar lo siguiente en la consola de python:

```
>>> import tkinter
>>> tkinter.Tcl().eval('info patchlevel') 
```

devolverá algo así

```
'8.6.8'
```

En el caso de que no esté instalado en linux deberemos instalar el paquete `python3_tk` con el gestor de paquetes propio de nuesto SO.

### Primera app tkinter - python

- Importaremos los módulos de tkinter necesarios, el original y la mejora de la versión 8 para controles avanzados.
- Definiremos los parámetros de la ventana en que se ejecutará la app
- Lanzaremos el loop de control de la ventana.
  - Esto es importante. Dado que las interfaces gráficas están orientada a eventos lanzarlas siempre equivale a un loop de espera a que se produzcan diferentes eventos. El loop finalizará cuando se produzca el evento de cierre de la ventana (en cada sistema operativo será pulsar un determinado botón de la ventana)
  
Veámoslo (tkinter no funciona en la red, copia el código y úsalo en tu IDE)

````
from tkinter import *
from tkinter import ttk
 
mainWindow = Tk()
 
mainWindow.geometry('640x480')
mainWindow.configure(bg = 'gray')
mainWindow.title('Aplicación')
 
mainWindow.mainloop()
````
