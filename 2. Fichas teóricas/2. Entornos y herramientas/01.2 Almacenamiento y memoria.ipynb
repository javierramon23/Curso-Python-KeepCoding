# Almacenamiento y memoria

## Memoria

Hemos visto en [como funciona un ordenador](https://colab.research.google.com/drive/1BDNigE3q4w-SQ9HDVgogE64WSszjoKJO#scrollTo=aZL50orPEghP) que uno de los elementos esenciales (aparte del procesador) para que trabaje el ordenador es la memoria.

La memoria no es más que un almacen de datos. Sin entrar en detalles profundos cualquier memoria de la que dispongamos es el equivalente a un casillero con celdas que pueden ser rellenadas con valores numéricos.

Estos números componiéndose con otros dan lugar a todos los[ tipos de datos](https://drive.google.com/open?id=1pcBNKNhKed3RjvcpsYfNPSIKdvfrt70P) que conocemos (primitivos o compuestos).

Un ordenador tiene dos tipos de memoria, la **ROM** (*Read Only Memory*) y la **RAM** (*Random Access Memory*). 

La primera (**ROM**) es una memoria permanente que sólo puede leerse y tiene lo necesario para poner en marcha el sistema (normalmente comprueba los accesos a los dispositivos de entrada/salida principales y busca en secuencia el sistema operativo en los distintos discos para que el ordenador acabe de encenderse). Lógicamente la ROM mantiene sus datos incluso con el ordenador apagado.

La RAM en cambio es volátil y mutable, es decir se borra cuando el ordenador se apaga y puede modificarse según se necesita. Aquí se cargan los programas que están en ejecución en un momento dado y todos sus datos. Los distintos ordenadores (sobremesas, laptops, tablets y móviles) que utilizamos cada vez consumen más RAM ya que cada vez son más potentes.

## Almacenamiento

Los programas y datos que se utilizan en un ordenador desaparecen de su memoria cuando se apaga (como todos sabemos por apagones inesperados), así que necesitamos dispositivos en los que almacenar la información mientras nuestro ordenador está desconectado.

Estas son las unidades de memoria externas, discos duros, pens usb,... En el fondo funcionan de igual manera, son casilleros en los que se deja información numérica, solo que al ser de forma magnética u óptica y no electrica, no necesitan corriente electrica para mantener la información.



## ¿Cómo accedemos a los datos almacenados que queremos?

Al principio las unidades de memoria tenían tan poca capacidad que se podía acceder por dirección al casillero, así si habíamos guardado un número entero en 4 celdas consecutivas a partir de la celda 127, podíamos acceder a esa información indicando algo así:
```
a = &127
```

De hecho es así como funciona el procesador con la memoria RAM y ROM. Sin embargo no es una buena forma de acceso a las unidades de almancenamiento masivo.
