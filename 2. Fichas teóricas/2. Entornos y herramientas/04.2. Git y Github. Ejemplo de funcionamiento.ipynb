# Git y Github. Ejemplo en marcha

## Requisitos

1. Crear cuenta en [github](https://github.com/).
2. Instalar git en local ([download](https://git-scm.com/downloads))

## Crear proyecto en local
1. Abrimos una consola 
2. Nos situamos en la carpeta donde vamos a realizar el proyecto
3. Ejecutamos
```
git init
git add .
git commit
```
`git init` inicializa el directorio como repositorio de git y crea la zona de trabajo [`git repo`](https://colab.research.google.com/drive/1xwcD9GAbEjiPZ6eVrTSQLU6A0npMvbUs#scrollTo=jYWMXoSm8hxR&line=25&uniqifier=1)
`git add .` pasa todo el contenido de nuestro directorio de trabajo a la staging area. Si no hay ningún fichero se puede crear el siguiente `readme.md` en el que incluiremos las instrucciones para ejecutar el proyecto.
`git commit` pasa los ficheros de la staging area a la zona repo git.

## Crear un proyecto en el repositorio central (github)

Utilizamos la interfaz gráfica de github para crear un repositorio nuevo:
- `New repository`
- Se le asigna un nombre y se crea como proyecto público (los privados tienen coste) y se hace click en `Create repository`

## Seguir las instrucciones para hacer push del repositorio local

Segundo ejemplo de la página de github de nuestro repositorio remoto en github
```
git remote add origin https://github.com/<nombre_de_usuario>/<nombre_de_proyecto>.git
git push -u origin master
```
Ya hemos hecho nuestro primer commit sobre nuestro proyecto y lo tenemos situado en el repositorio remoto.

## Clonar un proyecto en nuestra máquina

Si queremos clonar un proyecto de github en nuestra máquina (o disponemos de dos ordenadores y queremos bajarnos nuestro proyecto a otro ordenador).

Tomamos la url del proyecto remoto y tecleamos

```
git clone <url_remota>
````

Nos creará un directorio con el nombre del proyecto. Ahí podremos modificar lo que nos interese.

## Modificar y crear una nueva versión

Dentro del directorio de nuestro proyecto modificaremos archivos y crearemos otros. Cuando consideremos que ya es momento de hacer una nueva versión ejecutaremos
```
git status
git add .
git commit -m "Descripción detallada de la versión"
```
- `git status`: Muestra los archivos que están en staging area y los que no
- `git add .`: Situa todos los ficheros nuevos o modificados de nuestro directorio en la staging_area
- `commit -m ": Descripción detallada de la versión"`: Pasa todos los archivos de las staging_area a una nueva versión. La descripción es importante para luego entender que hace cada versión

## Subir nueva versión al repositorio remoto
Esta nueva versión solo figura en nuestro equipo. Para enviarla al repositorio remoto
```
git push origin master
````

## Comprobando nuestros commit
Se debe ejecutar la siguiente instrucción:
```
git log
```
Se ven los commits por orden
