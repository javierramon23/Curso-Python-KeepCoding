# Dispositivos de entrada/salida

Como ya hemos visto en la estructura de un ordenador, este se comunica con otros ordenadores o usuarios a través de sus dispositivos de entrada y salida.

Los dispositivos más habituales son:

* Entrada
  * Teclado
  * micrófono
  * pantalla táctil
  * cámara
  * sensores de movimiento
* Salida
  * Pantalla
  * Altavoces
  * Impresoras
  * Actuadores: Motores y otros dispositivos que actúan en respuesta a impulsos electrónicos (valvulas de inyección de combustible, o alerones en el caso del piloto automático de los aviones, por ejemplo)
  
Todos estos dispositivos deben ser controlados tanto en la lectura de datos (entrada) como en la presentación de los resultados (datos de salida) por nuestro programa.
