# Elementos de un programa

Recordamos que un [programa](https://drive.google.com/open?id=1zwpCGHNAHoC1DNo4MVHJ4fW-q_fQlSrX) es una receta para resolver un problema con un ordenador.

Una receta no es más que una secuencia de instrucciones que deben realizarse una detrás de otra. Vamos a analizar aquí como es esa secuencia de instrucciones.

Los programas se componen básicamente de tres elementos
  - [Instrucciones](https://drive.google.com/open?id=1wgw1ZrAOLfIitFrSNtair8aSdBXnQWHo)
  - [Estructuras de control](https://colab.research.google.com/drive/1Pfcyzl_UW6oc9U1CZQCXO0fj8h9cNN9a#scrollTo=ecNlILQxuFBe)
  - [Datos](https://drive.google.com/open?id=19zQfzOwWOsflSLS7h-MA8mbE9vszQA3T)
