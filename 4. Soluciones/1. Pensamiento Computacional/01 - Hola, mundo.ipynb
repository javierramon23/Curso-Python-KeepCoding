## Segundo Reto

```
nombre = input("¿Cómo te llamas? ")

if nombre == "Mon":
  print("Hola,", nombre, "¿Cómo estás? ")
elif nombre == "Pilar":
  print("Hola, Pili. Encantado de volver a verte")
elif nombre == "Belen":
  print("¿Qué pachaaaaa? ")
elif nombre == "Marisa":
  print("A las órdenes de vuecencia, doña Marisa")
else:
  print("Hola, ", nombre)
```
